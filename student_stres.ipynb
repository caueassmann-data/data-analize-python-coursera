```python
import pandas as pd 
```

```python
df=pd.read_csv('../Datasets_Kaggle/KaggleStudentStress/Stress_Dataset.csv')
df2=pd.read_csv('../Datasets_Kaggle/KaggleStudentStress/StressLevelDataset.csv')
print(df2.head())
```

```python
age_min=df['Age'].min()
age_max=df['Age'].max()
print(f'Min age: {age_min}, Max age: {age_max}')
```

```python
filtro = df['Age'].between(18, 30, inclusive='both')
soma_participantes_18_30 = filtro.sum()
print(f'Number of participants aged between 18 and 30: {soma_participantes_18_30}')
```

```python
n = df.query('18 <= Age <= 30')["Which type of stress do you primarily experience?"] \
      .str.contains('Distress', case=False, na=False) \
      .sum()
print(n)  # deve dar 30
```
